```python
import pandas as pd
```

```python
#Carrega os df adquiridos via web scraping
df_guarulhos = pd.read_csv("../dataset/guarulhos.csv")
df_pinheiros = pd.read_csv("../dataset/pinheiros.csv")
df_itaim_bibi = pd.read_csv("../dataset/itaim_bibi.csv")
df_santo_amaro = pd.read_csv("../dataset/santo_amaro.csv")
```

```python
#Adiciona a coluna 'região'
df_guarulhos['regiao'] = 'Centro Guarulhos'
df_itaim_bibi['regiao'] = 'Itaim Bibi'
df_pinheiros['regiao'] = 'Pinheiros'
df_santo_amaro['regiao'] = 'Santo Amaro'
```

```python
#Trata e cria uma coluna para cada atributo da coluna Conveniências
lista_df = [df_guarulhos, df_itaim_bibi, df_pinheiros, df_santo_amaro]

for df in lista_df:

    df['estrelas'] = (
        df['Conveniências']
        .str.extract(r'(\d)\s*estrela', expand=False)
    ).astype(float)


    df['wifi'] = df['Conveniências'].str.contains(
        'Wi-Fi', case=False, na=False
    ).astype(int)
    
    df['cafe_manha'] = df['Conveniências'].str.contains(
        'Café da manhã', case=False, na=False
    ).astype(int)

    df['piscina'] = df['Conveniências'].str.contains(
        'Piscina', case=False, na=False
    ).astype(int)

    df['estacionamento'] = df['Conveniências'].str.contains(
        'Estacionamento', case=False, na=False
    ).astype(int)

    df['academia'] = df['Conveniências'].str.contains(
        'Academia', case=False, na=False
    ).astype(int)

    df['spa'] = df['Conveniências'].str.contains(
        'Spa', case=False, na=False
    ).astype(int)

    df['ar_condicionado'] = df['Conveniências'].str.contains(
        'Ar-condicionado', case=False, na=False
    ).astype(int)

    df['pet_friendly'] = df['Conveniências'].str.contains(
        'Aceita animais de estimação', case=False, na=False
    ).astype(int)
```

```python
#Junta os df em um só, e exclui colunas que possuem informações não relevantes
df_final = pd.concat(lista_df)
df_final = df_final.drop(
    columns=['Nome do Hotel', 'Link', 'Conveniências']
)
```

```python
#Trata a coluna região para valor booleano utilizando .get_dummies, assumindo a região de 'Guarulhos' como caso base
df_modelo = pd.get_dummies(
    df_final,
    columns=['regiao'],
    drop_first=True
)
```

```python
#Trata valores nulos
df_modelo['estrelas'] = df_modelo['estrelas'].fillna(
    df_modelo['estrelas'].median()
)

df_modelo = df_modelo.dropna(subset=['Valor da Diária (R$)'])

df_modelo['Avaliação'] = df_modelo['Avaliação'].fillna(
    df_modelo['Avaliação'].median()
)

```

```python
#salva df_modelo para .csv
df_modelo.to_csv("../dataset/df_modelo.csv", index=False)
```
